```python
import cv2
import mediapipe as mp
import time


cap = cv2.VideoCapture(0) # Mudar o número para corresponder a camera (em geral é 0 ou 1, mas pode ser outro inteiro)
adress = 'http://192.168.0.129:8080/video' #Necessario quando usado webcam por IP
cap.open(adress)

mp_hands = mp.solutions.hands

# instanciando a classe
hands = mp_hands.Hands(
    static_image_mode=False,
    max_num_hands=1,
    min_detection_confidence=0.9,
    min_tracking_confidence=0.9
) # equivalente a mp.solutions.hands.Hands()

# Utilizando o drawing_utils para desenhar as mãos
mp_draw = mp.solutions.drawing_utils
DrawingSpec = mp.solutions.drawing_utils.DrawingSpec

while True:
    sucesso, img = cap.read()
    
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    results = hands.process(img_rgb)
    
    h, w, c = img.shape # h: altura, w: largura, c: canais de cor
    letra_tela = '-'
    if (results.multi_hand_landmarks):
        for hand_number, hand_landmark in enumerate(results.multi_hand_landmarks):
            tab_x = []
            tab_y = []
            tab_z = []
            for i, landmark in enumerate(hand_landmark.landmark):
                tab_x.append(int(landmark.x * w))
                tab_y.append(int(landmark.y * h))
                tab_z.append(int(landmark.z * 100))
            if (tab_y[8] > tab_y[5] and
                tab_y[12] > tab_y[9] and
                tab_y[16] > tab_y[13] and
                tab_y[20] > tab_y[17]
               ):
                letra_tela = 'A'
           
            if (tab_y[8] < tab_y[7] and
                tab_y[12] < tab_y[11] and
                tab_y[16] < tab_y[15] and
                tab_y[20] < tab_y[19] and
                tab_x[4] < tab_x[3]
               ):
                letra_tela = 'B'
            
            if (tab_x[8] > tab_x[6] and
                tab_x[12] > tab_x[10] and
                tab_x[16] > tab_x[14] and
                tab_x[20] > tab_x[18] and
                tab_x[4] > tab_x[2] and
                tab_z[5] > tab_z[17] and  
                tab_y[8] > tab_y[7] and
                tab_y[16] > tab_y[15] and
                tab_y[20] > tab_y[19]
               ):
                letra_tela = 'C'
                
            if (tab_x[8] > tab_x[6] and
                tab_x[12] > tab_x[10] and
                tab_x[16] > tab_x[14] and
                tab_x[20] > tab_x[18] and
                tab_x[4] > tab_x[2] and
                tab_z[5] > tab_z[17] and  
                tab_y[8] < tab_y[7] and
                tab_y[16] > tab_y[15] and
                tab_y[20] > tab_y[19]
               ):
                letra_tela = 'D'
                
            if (tab_y[8] < tab_y[5] and
                tab_y[12] < tab_y[9] and
                tab_y[16] < tab_y[13] and
                tab_y[20] < tab_y[17] and
                tab_y[8] > tab_y[7] and
                tab_y[12] > tab_y[11] and
                tab_y[16] > tab_y[15] and
                tab_y[20] > tab_y[19] and
                tab_x[4] < tab_x[3]
               ):
                letra_tela = 'E'
            
            if (tab_y[12] < tab_y[11] and
                tab_y[16] < tab_y[15] and
                tab_y[20] < tab_y[19] and
                tab_y[8] > tab_y[4]
               ):
                letra_tela = 'F'
                
            if (tab_y[8] < tab_y[7] and
                tab_y[12] > tab_y[9] and
                tab_y[16] > tab_y[15] and
                tab_y[20] > tab_y[19]
               ):
            mp_draw.draw_landmarks(
                img,
                hand_landmark,
                mp_hands.HAND_CONNECTIONS)
    
    cv2.imshow("Imagem", img)
    cv2.waitKey(1)
```
